### コーパスの利用
`lib/python-3.x/site-packages/music21/corpus`  
以下に楽曲データベースが用意されている。  
今回はモーツァルトのK.545を解析してみよう。

```python
piece = m21.corpus.parse('mozart/k545/movement1_exposition.xml')

# 楽譜表示（musescoreやFinaleなどがインストールされていることが条件）
piece.show()
```

```python
# Web上にあるデータを解析する場合
url = 'http://kern.ccarh.org/cgi-bin/ksdata?l=cc/bach/cello&file=bwv1007-01.krn&f=xml'

# 任意のxmlファイルをparseするときは　m21.converter.parse()を利用する。
bwv1007 = m21.converter.parse(url)

# 1小節目から5小節目までを表示
bwv1007.measures(1, 5).show()
```

```python
# midi で聞く場合
piece.show('midi')
```

```python
# 調を調べる（musicXMLの中に書いてある）
piece.analyze('key')
```

```python
# ピアノロール表示
piece.plot('pianoroll')
```

```python
# ヒストグラム
piece.plot('histogram')
```

```python
# 左手のピッチクラスだけヒストグラム
## ハ長調のヒストグラムになっていることがわかる
piece[2].plot('histogram','pitchclass')
```

```python
# 右手はほとんどが16分音符であることがわかる
piece[1].plot('histogram','quaterLength')
```

```python
piece.plot(format='scatterweighted', values='pitch')
```
